# 5.Visualize the Nobel Prize winners using their gender, age, award category and year. Interpret the image according to the results.

<font color='orange'>***DIVIDING THE YEAR COLUMN INTO 20-YEAR PERIODS***</font>

```python
def yirmi_yil(row, col_name):
    
    d = int(row[col_name])   
    
    if(d > 1900 and d < 1920):
        return "1900+"    
    elif(d >= 1920 and d < 1940):
        return "1920+"
    elif(d >= 1940 and d < 1960):
        return "1940+"
    elif(d >= 1960 and d < 1980):
        return "1960+"
    elif(d >= 1980 and d < 2000):
        return "1980+"
    elif(d >= 2000 and d < 2020):
        return "2000+"
    else:
        return 'Unknown'
nobel['Yirmi_Yil'] = nobel.apply (lambda row: yirmi_yil(row, 'year'),axis=1)
```

<font color='orange'>***FINDING GENDER DISTRIBUTIONS BY CATEGORIES***</font>

```python
sns.countplot(data=nobel, x=nobel.category,hue=nobel.sex)
plt.title("Kategorilere Göre Cinsiyet Dağılımları ")
plt.xlabel("Kategoriler")
plt.ylabel("Ödül Sayıları")
```

<font color='orange'>***FINDING THE GENDER DISTRIBUTION BY YEARS***</font>

```python
sns.countplot(data=nobel, x=nobel.Yirmi_Yil,hue=nobel.sex)
plt.title("20 Yıllık Periyotlara Göre Cinsiyet Dağılımları ")
plt.xlabel("Yıllar")
plt.ylabel("Ödül Sayıları")
```

<font color='orange'>***FINDING AGE DISTRIBUTIONS BY CATEGORIES***</font>

<font color='orange'>***AGE CALCULATION***</font>

```python
import datetime
nobel["dogum_yili"]=nobel.birth_date.dt.strftime("%Y")
```

```python
nobel['dogum_yili'] = pd.to_numeric(nobel['dogum_yili'])
```

```python
nobel["yas"] =nobel["year"]-nobel["dogum_yili"]
```

```python
sns.scatterplot(data=nobel, x=nobel.yas, y=nobel.category,color="orange")
plt.title("Kategorilere Göre Yaş Dağılımları ")
plt.xlabel("Yaşlar")
plt.ylabel("Kategoriler")
```

<font color='orange'>***FINDING AGE DISTRIBUTIONS BY YEARS***</font>

```python
sns.scatterplot(data=nobel, x=nobel.yas, y=nobel.Yirmi_Yil,color="orange")
plt.title("20 Yıllık Periyotlara Göre Yaş Dağılımları ")
plt.xlabel("Yaşlar")
plt.ylabel("Yıllar")
```
